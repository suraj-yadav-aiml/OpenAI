# [Top p - openai-ref.](https://platform.openai.com/docs/api-reference/completions/create#completions/create-top_p)
### Optional ,  Defaults to 1


````python
display(Image('img/5.png'))
display(Image('img/1.png'))
````

<p><font size="4">The <strong>"Top p"</strong> parameter, also known as the <strong>"nucleus"</strong> or <strong>"top probability"</strong> parameter, is a feature used in the ChatGPT API that helps control the randomness and diversity of the generated responses. It allows you to set a threshold probability value, and only the most probable tokens that cumulatively exceed this probability are considered during response generation.</font></p>
<p><font size="4">To explain it in simple terms, let's say the <strong>"Top p"</strong> parameter is set to <strong>0.8</strong>. During response generation, the model considers the probability distribution of the next possible tokens based on the context provided. It then calculates the cumulative probability of these tokens from the most probable to the least probable. It keeps adding up the probabilities until the sum exceeds the threshold set by the <strong>"Top p"</strong> value (<strong>0.8</strong> in this example). The model then samples from this narrowed-down set of tokens to choose the next word in the response.</font></p>
<p><font size="4">Here are five detailed examples to illustrate how the <strong>"Top p"</strong> parameter works:</font></p>
<ol>
  <li>
    <font size="4"><strong>Example with "Top p" set to 0.8:</strong><br>
    Suppose the model generates the following probabilities for the next tokens: [<strong>0.3</strong>, <strong>0.2</strong>, <strong>0.15</strong>, <strong>0.1</strong>, <strong>0.05</strong>, <strong>0.04</strong>, <strong>0.03</strong>]. The cumulative probabilities would be: [<strong>0.3</strong>, <strong>0.5</strong>, <strong>0.65</strong>, <strong>0.75</strong>, <strong>0.8</strong>, 0.84, 0.87]. Since the cumulative probability exceeds 0.8 after considering the first five tokens, only those tokens are eligible for sampling, and the model chooses from them.</font></li>
  <li>
    <font size="4"><strong>Example with "Top p" set to 0.5:</strong><br>
    Suppose the probabilities for the next tokens are: [<strong>0.4</strong>, <strong>0.25</strong>, <strong>0.2</strong>, 0.05, 0.03, 0.02, 0.01]. The cumulative probabilities would be: [<strong>0.4</strong>, <strong>0.65</strong>, <strong>0.85</strong>, 0.9, 0.93, 0.95, 0.96]. Here, the cumulative probability exceeds 0.5 after considering the first three tokens. Thus, the model only samples from those three tokens.</font></li>
  <li>
    <font size="4"><strong>Example with "Top p" set to 0.2:</strong><br>
    If the probabilities for the next tokens are: [<strong>0.1</strong>, <strong>0.09</strong>, 0.08, 0.07, 0.06, 0.05, 0.04], the cumulative probabilities would be: [<strong>0.1</strong>, <strong>0.19</strong>, <strong>0.27</strong>, 0.34, 0.4, 0.45, 0.49]. As the cumulative probability surpasses 0.2 after the second token, only the first two tokens are eligible for sampling.</font></li>
  <li>
    <font size="4"><strong>Example with "Top p" set to 1.0 (No truncation):</strong><br>
    If the probabilities for the next tokens are: [<strong>0.3</strong>, <strong>0.2</strong>, <strong>0.15</strong>, <strong>0.1</strong>, <strong>0.05</strong>, <strong>0.04</strong>, <strong>0.03</strong>], the cumulative probabilities would be: [<strong>0.3</strong>, <strong>0.5</strong>, <strong>0.65</strong>, <strong>0.75</strong>, <strong>0.8</strong>, 0.84, 0.87]. Since the cumulative probability exceeds 1.0 after considering all the tokens, no truncation occurs, and the model samples from all the tokens.</font></li>
  <li>
    <font size="4"><strong>Example with "Top p" set to 0.0 (No sampling):</strong><br>
    In this case, regardless of the probabilities assigned to the next tokens, the <strong>"Top p"</strong> value is 0.0, which means no tokens are considered for sampling. The model would always choose the most probable token as the next word, resulting in deterministic responses.</font></li>
</ol>
<p><font size="4">Remember, adjusting the <strong>"Top p"</strong> parameter can influence the randomness and creativity of the generated responses, allowing you to control the balance between coherent and unexpected outputs.</font></p>

#### OpenAI reference :
#### top_p :
An alternative to sampling with temperature, called nucleus sampling, where the model considers the results of the tokens with top_p probability mass. So 0.1 means only the tokens comprising the top 10% probability mass are considered.
**We generally recommend altering `top_p` or `temperature` but not both.**

````python
import textwrap
````

````python
def nice_print(dictionary):
    for key, value in dictionary.items():
        print(f"{key}:")
        print("\n".join(textwrap.wrap(value,width=120)))
        print("-"*120)
````

````python
nice_print(
    {
        f"Top p {top_p}": openai.Completion.create(
            model = 'text-davinci-003',
            prompt = """My favorite food is""".strip(),
            max_tokens = 50,
            echo = True,
            top_p=top_p
        )['choices'][0]['text']
        .strip()
        for top_p in [1, 0.5, 0]
    }
)
````

````python
def nice_print_2(dictionary):
    for key, value in dictionary.items():
        print(f"{key}:")
        print(textwrap.fill(value,width=120))
        print("-"*120)
````

````python
nice_print_2(
    {
        f"Top p {top_p}": openai.Completion.create(
            model = 'text-davinci-003',
            prompt = """My favorite food is""".strip(),
            max_tokens = 50,
            echo = True,
            top_p=top_p
        )['choices'][0]['text']
        .strip()
        for top_p in [1, 0.5, 0]
    }
)
````

# [frequency_penalty - OpenAI Ref.](https://platform.openai.com/docs/api-reference/completions/create#completions/create-frequency_penalty)
### number  ,   Optional  ,  Defaults to 0
```
Number between -2.0 and 2.0. Positive values penalize new tokens based on their existing frequency in the text so far, decreasing the model's likelihood to repeat the same line verbatim.
```
[See more information about frequency and presence penalties.](https://platform.openai.com/docs/api-reference/parameter-details)

````python
display(Image('img/6.png'))
display(Image('img/7.png'))
````

```
The "frequency_penalty" parameter in ChatGPT is used to control the repetitiveness of responses generated by the model. It encourages or discourages the model from repeating the same or similar phrases in its output. By adjusting the frequency penalty value, you can influence how likely the model is to generate repetitive responses.
```

<font size="4">
<p>
- The "<strong>frequency_penalty</strong>" parameter is a setting that you can use when interacting with ChatGPT to control the generation of repetitive or overly frequent responses.<br><br>
- It helps to adjust the model's tendency to repeat certain phrases or responses frequently, making the generated conversation more varied and diverse.<br><br>
- The value of "<strong>frequency_penalty</strong>" can range from -2 to 2. A lower value like -2 encourages the model to produce more repetitive responses, while a higher value like 2 discourages repetition and encourages more unique and novel responses.<br><br>
- A value of 0 (default) indicates no special preference for or against repetitive responses, and the model will generate based on its inherent biases.<br><br>
- A positive value above 0 increases the penalty on repetitive responses, making them less likely to be generated. This can help in situations where you want more originality in the model's output.<br><br>
- Conversely, a negative value below 0 decreases the penalty on repetitive responses, making them more likely to occur. This can be useful if you want the model to be more consistent or if repetition is acceptable or desired in the context.<br>
- The actual impact of the "<strong>frequency_penalty</strong>" setting can vary depending on the specific conversation and the other parameters used. It may not completely eliminate repetition, but it can influence the likelihood of repetitive responses.<br><br>
- It's important to experiment with different values of "<strong>frequency_penalty</strong>" to find the balance that suits your desired output and the nature of the conversation you're having with ChatGPT.
</p>
</font>

````python
def nice_print(dictionary):
    for key, value in dictionary.items():
        print(f"{key}:")
        print(textwrap.fill(value,width=120))
        print("-"*120)
````

````python
nice_print(
    {
        f"Frequency Penalty {x}": openai.Completion.create(
            model="text-davinci-003",
            prompt="""The first 15 elements are Hydrogen, Helium,""".strip(),
            max_tokens=200,
            echo=True,
            frequency_penalty=x,
        )['choices'][0]["text"]
        .strip()
        for x in [-2, -1, 0, 1, 2]
    }
)
````

## What is difference between presence_penalty and frequency_penalty ?

<p><span style="font-size: 20px;">The <strong>"presence_penalty"</strong> and <strong>"frequency_penalty"</strong> are two parameters in the ChatGPT API that affect the diversity and repetitiveness of model responses, but they operate in different ways. Here's a comparison of the two:</span></p>

<h3><span style="font-size: 20px;">Presence Penalty:</span></h3>
<ul>
<li><span style="font-size: 20px;">Controls the model's tendency to generate repetitive or redundant responses.</span></li>
<li><span style="font-size: 20px;">Values range from <strong>-2 to 2</strong>.</span></li>
<li><span style="font-size: 20px;">A higher positive value (e.g., <strong>2</strong>) encourages the model to produce more diverse responses, reducing repetition.</span></li>
<li><span style="font-size: 20px;">A negative value (e.g., <strong>-2</strong>) promotes more focused and repetitive responses.</span></li>
<li><span style="font-size: 20px;">A value of <strong>0</strong> has no effect on the model's behavior in terms of repetition.</span></li>
</ul>

<h3><span style="font-size: 20px;">Frequency Penalty:</span></h3>
<ul>
<li><span style="font-size: 20px;">Controls the model's sensitivity to using words or phrases that have been recently used in the conversation.</span></li>
<li><span style="font-size: 20px;">Values range from <strong>-2 to 2</strong>.</span></li>
<li><span style="font-size: 20px;">A higher value (e.g., <strong>2</strong>) makes the model more cautious about repeating words or phrases used earlier in the conversation.</span></li>
<li><span style="font-size: 20px;">A value of <strong>0</strong> has no effect on the model's behavior regarding word frequency.</span></li>
<li><span style="font-size: 20px;">Higher values like <strong>0.8</strong> or <strong>1.0</strong> are commonly used to minimize repeated terms.</span></li>
</ul>

<p><span style="font-size: 20px;">In summary, the key differences between <strong>presence_penalty</strong> and <strong>frequency_penalty</strong> are as follows:</span></p>
<ul>
<li><span style="font-size: 20px;">Presence_penalty focuses on avoiding repetitive overall responses, encouraging diversity in the generated output.</span></li>
<li><span style="font-size: 20px;">Frequency_penalty specifically targets the repetition of specific words or phrases within the conversation, aiming to reduce redundancy.</span></li>
</ul>

<p><span style="font-size: 20px;">Both parameters contribute to making the model's responses more varied and less repetitive, but they operate at different levels. Presence_penalty influences the overall response diversity, while frequency_penalty specifically addresses the repetition of words or phrases within the conversation.</span></p>

# [stream](https://platform.openai.com/docs/api-reference/completions/create#completions/create-stream)

### boolean ,  Optional  ,  Defaults to false

```
Whether to stream back partial progress. If set, tokens will be sent as data-only server-sent events as they become available, with the stream terminated by a data: [DONE] message
```

```
The "stream" parameter in the Completion API allows you to receive a stream of responses from the model instead of a single response. This can be useful when you want to have an interactive conversation with the model in real-time, where you send multiple messages and receive immediate responses. Here's a detailed explanation of the "stream" parameter:
```
 1. The "stream" parameter is a boolean value that determines whether the API response should be streamed or not.
 2. When "stream" is set to true, the API will continuously send responses in a streaming fashion until the completion is finished or an error occurs.

````python
response = openai.Completion.create(model='text-davinci-003',
                                    prompt = "Write me a poem about Python Developer",
                                    max_tokens=300)

print(response['choices'][0]['text'])
````

````python
response = openai.Completion.create(model='text-davinci-003',
                                    prompt = "Write me a poem about Python Developer",
                                    max_tokens=300,
                                    stream=True)

for data in response:
    print(data['choices'][0]['text'], end="")
````

````python
response = openai.Completion.create(model='text-davinci-003',
                                    prompt = "Write me a poem about Python Developer",
                                    max_tokens=300,
                                    stream=True)

for data in response:
    print(data['choices'][0]['text'], end="", flush=True)
````
